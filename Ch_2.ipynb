## Data Structures 
### Shared Vocab

```python
nlp.vocab.strings.add("coffee")
coffee_hash = nlp.vocab.strings["coffee"]
coffee_string = nlp.vocab.strings[coffee_hash]
```

```python
print("hash value:", coffee_hash)
print("string value:", nlp.vocab.strings[coffee_hash])
```

```python
doc = nlp("I love coffee")
print("hash value:", doc.vocab.strings["coffee"])
```

```python
for i in doc.vocab.strings:
    print(i)
```

### Lexemes: Entries in the vocabulary
- contains context-independent information about a word
- can be obtained by looking up a string or hash ID in the vocab

```python
lexeme = nlp.vocab["coffee"]
```

```python
print(lexeme.text, lexeme.orth, lexeme.is_alpha)
```

### Doc, Span, Token
- Doc and Span hold references and relationships of words and sentences


#### 1. Doc
- Doc class takes three arguments: shared vocab, words, spaces

```python
nlp = spacy.blank("en")
from spacy.tokens.doc import Doc
```

```python
# Manually create a doc
words = ["Hello", "world", "!"]
spaces = [True, False, False]
doc = Doc(nlp.vocab, words=words, spaces=spaces)
doc
```

#### 2. Span
- slice of a doc
- consists of one or more tokens

```python
from spacy.tokens import Span
```

```python
span = Span(doc, 0, 2, label="GREETING")
span
```

```python
# add span to the doc.ents
doc.ents = [span]
doc.ents
```

#### Tips
- convert results to strings as late as possible
- use token attributes if available (e.g. token.i for token index)
