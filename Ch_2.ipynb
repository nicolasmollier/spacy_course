### Doc, Span, Token
- Doc and Span hold references and relationships of words and sentences


#### 1. Doc
- Doc class takes three arguments: shared vocab, words, spaces

```python
nlp = spacy.blank("en")
from spacy.tokens.doc import Doc
```

```python
# Manually create a doc
words = ["Hello", "world", "!"]
spaces = [True, False, False]
doc = Doc(nlp.vocab, words=words, spaces=spaces)
doc
```

#### 2. Span
- slice of a doc
- consists of one or more tokens

```python
from spacy.tokens import Span
```

```python
span = Span(doc, 0, 2, label="GREETING")
span
```

```python
# add span to the doc.ents
doc.ents = [span]
doc.ents
```

#### Tips
- convert results to strings as late as possible
- use token attributes if available (e.g. token.i for token index)
- pass in the shared vocab

```python
# Bad example
nlp = spacy.load("en_core_web_sm")
doc = nlp("Berlin looks like a nice city")

# Get all tokens and part-of-speech tags
token_texts = [token.text for token in doc]
pos_tags = [token.pos_ for token in doc]

for index, pos in enumerate(pos_tags):
    # Check if the current token is a proper noun
    if pos == "PROPN":
        # Check if the next token is a verb
        if pos_tags[index + 1] == "VERB":
            result = token_texts[index]
            print("Found proper noun before a verb:", result)
```

```python
# Good example: Use native token attributes instead of lists of strings
nlp = spacy.load("en_core_web_sm")
doc = nlp("Berlin looks like a nice city")

# Iterate over the tokens
for token in doc:
    # Check if the current token is a proper noun
    if token.pos_ == "PROPN":
        # Check if the next token is a verb
        if doc[token.i + 1].pos_ == "VERB":
            print("Found proper noun before a verb:", token.text)
```

## Word Vectors and semantic similarity

How does spaCy predict similarity?

- word vectors (e.g. word2vec)
- default: cosine similarity
- `Doc` and `Span` object default to average of token vectors

Similarity depends on the application context

- applications: recommendation systems, flagging duplicates

```python
%%capture
!python -m spacy download en_core_web_md
```

```python
# Load a larger pipeline with vectors
nlp = spacy.load("en_core_web_md")

# Compare two documents
doc1 = nlp("I like fast food")
doc2 = nlp("I like pizza")
print(doc1.similarity(doc2))
```

```python
# Compare two tokens
doc = nlp("I like pizza and pasta")
token1 = doc[2]
token2 = doc[4]
print(token1.similarity(token2))
```

```python
# Compare a document with a token
doc = nlp("I like pizza")
token = nlp("soap")[0]

print(doc.similarity(token))
```

```python
# Compare a span with a document
span = nlp("I like pizza and pasta")[2:5]
doc = nlp("McDonalds sells burgers")

print(span.similarity(doc))
```

```python
doc = nlp("I have a banana")
print(f"Vector: \n{doc[3].vector}, \n\n Vector length: {len(doc[3].vector)}")
```

```python
# Similarity depends on the application context
doc1 = nlp("I like cats")
doc2 = nlp("I hate cats")

print(doc1.similarity(doc2))
```
